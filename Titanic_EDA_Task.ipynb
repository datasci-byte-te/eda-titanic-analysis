```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Optional: Set visual style
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("/content/Titanic-Dataset.csv")

 # change this to your file path
df.head()
```

```python
# Basic info
df.info()

# Shape of the dataset
print("Shape:", df.shape)

# Column names
print("Columns:", df.columns.tolist())

# Check missing values
df.isnull().sum()
```

```python
# Summary for numeric columns
df.describe()

# Summary for categorical columns
df.describe(include='object')

# or include='category'

```

```python
# Histogram of Age
sns.histplot(df['Age'].dropna(), kde=True)
plt.title("Distribution of Age")
plt.show()

# Boxplot of Fare
sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.show()


```

```python
# Histogram of SibSp - shows how many passengers had 0, 1, 2, ... siblings/spouses aboard
sns.histplot(df['SibSp'], bins=8, kde=False)
plt.title("Distribution of SibSp (Siblings/Spouses Aboard)")
plt.xlabel("SibSp")
plt.ylabel("Count")
plt.show()

# Boxplot for SibSp - helps identify outliers or concentration of values
sns.boxplot(x='SibSp', data=df)
plt.title("Boxplot of SibSp")
plt.show()




```

```python
# Histogram of Parch - shows how many passengers had 0, 1, 2, ... parents/children aboard
sns.histplot(df['Parch'], bins=8, kde=False)
plt.title("Distribution of Parch (Parents/Children Aboard)")
plt.xlabel("Parch")
plt.ylabel("Count")
plt.show()

# Boxplot for Parch - helps identify outliers or unusual distribution
sns.boxplot(x='Parch', data=df)
plt.title("Boxplot of Parch")
plt.show()




```

```python
# Histogram of Fare - shows the fare distribution, likely right-skewed
sns.histplot(df['Fare'], bins=40, kde=True)
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()

# Boxplot of Fare - helps detect extreme fare values (outliers)
sns.boxplot(x='Fare', data=df)
plt.title("Boxplot of Fare")
plt.show()

```

```python
# Count plot for Sex
sns.countplot(x='Sex', data=df)
plt.title("Count of Sex")
plt.show()

# Survival rate by Pclass
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()




```

```python
# Interactive histogram of Fare
fig = px.histogram(df, x="Fare", color="Survived", nbins=30, title="Fare Distribution by Survival")
fig.show()

# Interactive scatter plot
fig = px.scatter(df, x="Age", y="Fare", color="Survived", title="Age vs Fare by Survival")
fig.show()




```

```python
# Missing values again
print(df.isnull().sum())

# Outliers using boxplot
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("Outlier Detection")
plt.show()


```
